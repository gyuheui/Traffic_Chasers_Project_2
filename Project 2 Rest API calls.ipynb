```python
traffic_info = response["TRAFFIC_ITEMS"]
traffic_access= traffic_info['TRAFFIC_ITEM']
print(len(traffic_access))
```

```python
traffic_info = response["TRAFFIC_ITEMS"]

#print(traffic_info)
traffic_access= traffic_info['TRAFFIC_ITEM']
#print(traffic_access)

Traffic_type=[]
TrafficID = []
Criticality = []
Lat_Lon=[]
Start_time=[]
End_Time=[]
Lat = []
Lon=[]
CritID=[]
CritDescr=[]
ItemDescr=[]
CityName = []

CityNameChoices = {Atl_proximity_coords: "Atlanta, GA", NYC_proximity_coords: "New York City, NY"}
CityNameResult = CityNameChoices.get(proximity_coords, 'default')
print(CityNameResult)

CityAbrvChoices = {Atl_proximity_coords: "ATL", NYC_proximity_coords: "NYC"}
CityAbrvResult = CityAbrvChoices.get(proximity_coords, 'default')
print(CityAbrvResult)


i=0
for key in traffic_access:
    #print("KEY: "+ str(key))
    if i< len(traffic_access):
        Traffic_type.append(traffic_access[i]['TRAFFIC_ITEM_TYPE_DESC'])
        TrafficID.append(pd.to_numeric(traffic_access[i]['TRAFFIC_ITEM_ID']))
        Criticality.append(traffic_access[i]['CRITICALITY'])
        CritID.append(traffic_access[i]['CRITICALITY']['ID'])
        CritDescr.append(traffic_access[i]['CRITICALITY']['DESCRIPTION'])
        Lat_Lon.append(traffic_access[i]['LOCATION']['GEOLOC']['ORIGIN'])
        Lat.append(Lat_Lon[i]['LATITUDE'])
        Lon.append(Lat_Lon[i]['LONGITUDE'])
        Start_time.append(traffic_access[i]['START_TIME'])
        End_Time.append(traffic_access[i]['END_TIME'])
        ItemDescr.append(traffic_access[i]['TRAFFIC_ITEM_DESCRIPTION'][0]['value'])
        CityName.append(CityNameResult)
    i+=1
      

```

```python
#print(Traffic_type)
#print(Lat_Lon
print(ItemDescr)
print(CritDescr)
print(CritID)
print(Criticality)
#print(Start_time)
#print(End_Time)
```

```python
new_results_df=pd.DataFrame({"Traffic_Item_ID":TrafficID,
   "Traffic_types":Traffic_type,
    "Lat/Lon":Lat_Lon,
    "Lat":Lat,
    "Lon":Lon,
    "Start_time": Start_time,
    "End_Time": End_Time,
    "Criticality": Criticality,
    "CritID": CritID,
    "CritDescr": CritDescr,
    "ItemDescr": ItemDescr,
    "CityName": CityName
})
new_results_df
```

```python
print(datetime.now().strftime('%Y%m%d'))

df_new_results = new_results_df.astype(dtype={'Traffic_Item_ID': 'int64'}, copy=True)

df_new_results.to_csv(datetime.now().strftime('%Y%m%d')+"_" + CityAbrvResult + "_incidents.csv")
```
